```python
# load data
data = np.genfromtxt('train.csv', delimiter=',')
data = data[1:,:] # ignore header

X_final = np.genfromtxt('test.csv', delimiter=',')
X_final = X_final[1:,:]
```

```python
from sklearn.cross_validation import train_test_split
#split data into train, CV, test
train_set, test_set = train_test_split(data, test_size=0.50, random_state=42)
train_set, val_set = train_test_split(train_set, test_size=0.30, random_state=42)



print 'Training set   : %d data points' % len(train_set)
print 'Validation set : %d data points' % len(val_set)
print 'Test set       : %d data points' % len(test_set)

# split into input (X) and output (Y) variables
y_train = train_set[:,0:1]
X_train = train_set[:,1:]

y_val = val_set[:,0:1]
X_val = val_set[:,1:]

y_test = test_set[:,0:1]
X_test = test_set[:,1:]

y_all = data[:,0:1]
X_all = data[:,1:]
```

```python
def train_NN(X, Y, nb_epoch, batch_size, dim_layer1, dim_layer2):
    from keras.preprocessing import sequence
    from keras.models import Sequential
    from keras.layers import Dense, Dropout, Activation, Flatten
    from keras.layers import Embedding
    from keras.layers import Convolution1D, MaxPooling2D
    from keras.datasets import imdb
    from keras import backend as K
    import numpy
    # Create first network with Keras
    #from keras.models import Sequential
    #from keras.layers import Dense
    #from keras.layers import Pooling
    #import keras
    #import numpy
    # fix random seed for reproducibility
    seed = 7
    numpy.random.seed(seed)
    num_labels = 10
    Y_c = Y*np.ones([1,num_labels]) == [np.linspace(0,9,10)]
    
    
    # create model
    model = Sequential()
    model.add(Dense(dim_layer1, input_dim=784, init='lecun_uniform', activation='relu'))
    
    model.add(MaxPooling2D(pool_size=(4, 0), strides=None, border_mode='valid', dim_ordering='default'))
    #keras.layers.pooling.MaxPooling2D(pool_size=(2, 2), strides=None, border_mode='valid', dim_ordering='default')
    
    
    model.add(Dense(dim_layer2, init='lecun_uniform', activation='relu'))

    model.add(Dense(10, init='lecun_uniform', activation='sigmoid'))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    # Fit the model
    model.fit(X, Y_c, nb_epoch=nb_epoch, batch_size=batch_size, verbose=False)
    # evaluate the model
    scores = model.evaluate(X, Y_c)
    print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
    
    return model
```

```python
model = train_NN(X_train, y_train, nb_epoch=10, batch_size = 50, dim_layer1 = 25, dim_layer2 = 25)
```
